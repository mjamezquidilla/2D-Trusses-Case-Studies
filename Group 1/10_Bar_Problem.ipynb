## 10 Bar Problem

### Case 1

```python
from Truss_2D import Truss_2D
```

```python
nodes = {1: [720,360], 2:[720,0], 3:[360,360], 4:[360,0], 5:[0,360], 6:[0,0]}
elements = {1:[5,3], 2:[3,1], 3:[6,4], 4:[4,2], 5:[3,4], 6:[1,2], 7:[5,4], 8:[6,3], 9:[3,2], 10:[1,4]}
supports = {5:[1,1], 6:[1,1]}
forces = {2:[0,-100], 4:[0,-100]}
elasticity = {key: 10_000 for key in elements}
cross_area = {1:30, 2:1.62, 3:22.9, 4:13.5, 5:1.62, 6:1.62, 7:7.97, 8:26.5, 9:22, 10:1.8}
```

```python
Ten_Bar_Truss = Truss_2D(nodes = nodes,
                        elements= elements,
                        supports= supports,
                        forces = forces,
                        elasticity= elasticity,
                        cross_area= cross_area)
```

```python
Ten_Bar_Truss.Draw_Truss_Setup(figure_size=[10,6], length_of_arrow=50, width_of_arrow=10)
```

```python
Ten_Bar_Truss.Solve()
```

```python
Ten_Bar_Truss.member_forces_
```

```python
Ten_Bar_Truss.member_stresses_
```

```python
Ten_Bar_Truss.displacements_
```

```python
Ten_Bar_Truss.reactions_
```
